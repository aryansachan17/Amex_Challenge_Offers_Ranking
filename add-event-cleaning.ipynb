```python
df=pd.read_parquet('/kaggle/input/bitblenders-007/add_event.parquet')
df
```

```python
df = df.replace({None: np.nan})
df
```

```python
df['id4'] = pd.to_datetime(df['id4'])
df
```

```python
print(df.columns)
```

```python
print(f"--- Verification for 'id6' column ---")

# 1. Check unique categories and their counts
print("\nUnique values and their counts in 'id6':")
print(df['id6'].value_counts(dropna=False)) # dropna=False to also count NaN/None

# 2. Check for missing values
print(f"\nNumber of missing values in 'id6': {df['id6'].isnull().sum()}")

# 3. Check the data type
print(f"\nData type of 'id6': {df['id6'].dtype}")
```

```python


print(f"--- Missing value check for 'id2' column ---")

# 1. Count the number of missing values in 'id2'
missing_id2_count = df['id2'].isnull().sum()
print(f"\nNumber of missing values in 'id2': {missing_id2_count}")

# 2. Check the percentage of missing values in 'id2'
total_rows = len(df)
missing_id2_percentage = (missing_id2_count / total_rows) * 100
print(f"Percentage of missing values in 'id2': {missing_id2_percentage:.2f}%")

# 3. Display rows where 'id2' is missing (if any)
if missing_id2_count > 0:
    print("\nRows where 'id2' is missing:")
    print(df[df['id2'].isnull()])
else:
    print("\nNo missing values found in 'id2'.")
```
