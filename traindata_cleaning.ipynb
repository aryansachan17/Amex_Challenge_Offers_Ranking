```python
import pandas as pd
import os

train_df = pd.read_parquet('/kaggle/input/bitblenders-007/train_data.parquet')
train_df
```

```python
# Load metadata columns (id1 to id5, and target y)
meta_cols = ['id1', 'id2', 'id3', 'id4', 'id5', 'y']
df_meta = pq.read_table(base_path + 'train_data.parquet', columns=meta_cols).to_pandas()

# Convert to datetime
df_meta['id4'] = pd.to_datetime(df_meta['id4'], errors='coerce')
df_meta['id5'] = pd.to_datetime(df_meta['id5'], errors='coerce')

# Save cleaned metadata
df_meta.to_parquet(save_path + "train_meta_cleaned.parquet", index=False)
print("✅ Metadata cleaned and saved.")
```

```python
# Loop in batches of 50
for i in range(1, 367, 50):  # f1 to f366
    cols = [f"f{j}" for j in range(i, min(i+50, 367))]
    print(f"🔄 Cleaning columns: {cols[0]} to {cols[-1]}")

    # Load batch
    df_chunk = pq.read_table(base_path + 'train_data.parquet', columns=cols).to_pandas()

    # Replace invalid values with NaN
    df_chunk.replace(['None', None], np.nan, inplace=True)
    df_chunk.replace(-9999.0, np.nan, inplace=True)

    # Convert to numeric (force strings to become NaN)
    df_chunk = df_chunk.apply(pd.to_numeric, errors='coerce')

    # Save this cleaned batch
    out_path = save_path + f"train_clean_f{i}_{i+49}.parquet"
    df_chunk.to_parquet(out_path, index=False)
    print(f"✅ Saved: {out_path}")
```

```python
# Load all cleaned feature chunks
cleaned_chunks = []

for i in range(1, 367, 50):
    path = f"/kaggle/working/train_clean_f{i}_{i+49}.parquet"
    df_chunk = pd.read_parquet(path)
    cleaned_chunks.append(df_chunk)

# Combine features
df_features = pd.concat(cleaned_chunks, axis=1)

# Merge with cleaned metadata
df_meta = pd.read_parquet("/kaggle/working/train_meta_cleaned.parquet")
df_final = pd.concat([df_meta, df_features], axis=1)

# Save full cleaned data
df_final.to_parquet("/kaggle/working/train_data_cleaned_full.parquet", index=False)
print("🎉 Full cleaned dataset ready!")
```

```python
non_numeric_cols = df_final.select_dtypes(include=['object']).columns
print("Non-numeric columns:", non_numeric_cols)
```

```python
for col in non_numeric_cols:
    df_final[col] = df_final[col].str.replace(r'[^\d.-]', '', regex=True)  # Remove commas, slashes, etc.
    df_final[col] = pd.to_numeric(df_final[col], errors='coerce')
```

```python
non_numeric_cols = df_final.select_dtypes(include=['object']).columns
print("Non-numeric columns after cleaning:", non_numeric_cols)
```

```python
import re

def check_for_unwanted_chars(df, chars=r"[,/\\%]"):
    bad_cols = []
    for col in df.columns:
        if df[col].dtype == 'object':
            if df[col].str.contains(chars, regex=True).any():
                bad_cols.append(col)
    return bad_cols

unwanted = check_for_unwanted_chars(df_final)
print("Columns with unwanted characters:", unwanted)
```

```python
# Drop columns that have only 1 unique non-null value
nunique = df_final.nunique(dropna=True)
constant_cols = nunique[nunique <= 1].index.tolist()

print("Constant columns to drop:", constant_cols)

df_final.drop(columns=constant_cols, inplace=True)
```

```python
def cap_outliers(df, cols, lower_q=0.01, upper_q=0.99):
    for col in cols:
        if df[col].nunique() <= 1:
            continue
        lower = df[col].quantile(lower_q)
        upper = df[col].quantile(upper_q)
        df[col] = df[col].clip(lower, upper)
    return df

# Apply only to numerical feature columns (starting with 'f')
feature_cols = [col for col in df_final.columns if col.startswith('f')]
df_final = cap_outliers(df_final, feature_cols)

print("✅ Outliers capped.")
```

```python
df_final.fillna(df_final.median(numeric_only=True), inplace=True)
```
